```python
dots=sns.load_dataset('dots').query("align=='dots'")
```

```python
dots.head()
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',hue='coherence')
```

```python
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',hue='coherence',style='choice')
#styling is based on catogorical values
```

```python
palette=sns.cubehelix_palette(light=0.5,n_colors=6)
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',hue='coherence',style='choice',palette=palette)
```

```python
#To change the width of above lines
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',size='coherence',style='choice')
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',size='coherence',style='choice',sizes=(1,5))
sns.relplot(x='time',y='firing_rate',data=dots,kind='line',size='coherence',hue='coherence',style='choice',sizes=(1,5))
```
